## Merging dataframes together

The dataframes containing the median pIC50 values for each compound can be merged with the pairs dataframe, using the names of the compounds. This can be done in several merging steps. For this, the median files must first be imported.

# Calculating the change in activity between each pair of compounds

Calculating the change in activity that occurs after each transformation is simple; the pIC50 of B is just taken away from the pIC50 of A.

```python
egfr_sp_merge_ab["delta_pIC50"] = egfr_sp_merge_ab["pIC50 Value_A"] - egfr_sp_merge_ab["pIC50 Value_B"]
egfr_sp_merge_ab
```

```python
egfr_p_merge_ab["delta_pIC50"] = egfr_p_merge_ab["pIC50 Value_A"] - egfr_p_merge_ab["pIC50 Value_B"]
egfr_p_merge_ab
```

# SMIRK Analysis

Each transformation is denoted by a SMIRK. These SMIRKS can be grouped, to see how many examples of each transformation there are in the dataset. This will show which are the most common transformations between pairs in these datasets. 

The mean change in activity for each transformation can also be found. This shows which transformations cause the largest changes in activity on average. 

### Single Protein Assays:

```python
egfr_sp_1_count = egfr_sp_1['smirks'].value_counts()
egfr_sp_1_smirks = egfr_sp_1.groupby('smirks') [['delta_pIC50']].mean()

egfr_sp_1_merged = pd.merge(egfr_sp_1_smirks, egfr_sp_1_count, left_on="smirks", right_index=True)
egfr_sp_1_merged.rename(columns={egfr_sp_1_merged.columns[1]: 'Smirks_count'},inplace=True)

egfr_sp_2_count = egfr_sp_2['smirks'].value_counts()
egfr_sp_2_smirks = egfr_sp_2.groupby('smirks') [['delta_pIC50']].mean()

egfr_sp_2_merged = pd.merge(egfr_sp_2_smirks, egfr_sp_2_count, left_on="smirks", right_index=True)
egfr_sp_2_merged.rename(columns={egfr_sp_2_merged.columns[1]: 'Smirks_count'},inplace=True)

egfr_sp_3_count = egfr_sp_3['smirks'].value_counts()
egfr_sp_3_smirks = egfr_sp_3.groupby('smirks') [['delta_pIC50']].mean()

egfr_sp_3_merged = pd.merge(egfr_sp_3_smirks, egfr_sp_3_count, left_on="smirks", right_index=True)
egfr_sp_3_merged.rename(columns={egfr_sp_3_merged.columns[1]: 'Smirks_count'},inplace=True)

egfr_sp_4_count = egfr_sp_4['smirks'].value_counts()
egfr_sp_4_smirks = egfr_sp_4.groupby('smirks') [['delta_pIC50']].mean()

egfr_sp_4_merged = pd.merge(egfr_sp_4_smirks, egfr_sp_4_count, left_on="smirks", right_index=True)
egfr_sp_4_merged.rename(columns={egfr_sp_4_merged.columns[1]: 'Smirks_count'},inplace=True)
```

The first set of dataframes show the number of each transformation that is present in the datset.

These results show that each environment group actually contain different pairs, as the most commonly occurring SMIRKs are different in each group! This is important, as later analysis will therefore have to be performed on each environment group.

The second set of dataframes similarly shows the average change in ∆pIC50 for each transformation.

These dataframes can be merged together, so the average change in activity can be viewed alongside how many examples of each transformation there is in the set of pairs produced by MMPA. 

Ideally, there would be a group of around 10-20 examples of each transformation to provide a reasonable approximation of the average change in ∆pIC50, as anything less could simply be experimental error.
